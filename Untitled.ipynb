# Data Cleaning & Preprocessing

```python
df.columns
```

```python
# Step 1: Drop Unnecessary Columns
drop_cols = ['encounter_id', 'patient_nbr', 'weight', 'payer_code', 'medical_specialty']
df.drop(columns=drop_cols, inplace=True)
```

```python
# Step 2: Handle Missing Values (if any)
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
cols = ['num_procedures', 'number_outpatient', 'number_emergency', 'number_inpatient']
for col in cols:
    sns.histplot(df[col], bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Step 4: Encoding Categorical Features
cat_cols = df.select_dtypes(include='object').columns.tolist()
cat_cols
```

```python
# Label Encoding for binary columns
le = LabelEncoder()
for col in cat_cols:
    if df[col].nunique() == 2:
        df[col] = le.fit_transform(df[col])
```

```python
# One-Hot Encoding for multi-category columns
df = pd.get_dummies(df, columns=[col for col in cat_cols if df[col].nunique() > 2])
```

```python
df.head()
```

```python
df.dtypes
```

```python
bool_cols = df.select_dtypes(include='bool').columns.tolist()

# Object type columns with 2 unique values (binary categorical)
binary_obj_cols = [col for col in df.select_dtypes(include='object') if df[col].nunique() == 2]
binary_obj_cols
```

```python
for col in bool_cols:
    df[col] = df[col].astype(int)
```

```python
for col in binary_obj_cols:
    df[col] = df[col].map({df[col].unique()[0]: 0, df[col].unique()[1]: 1})
```

## Dimensionality Reduction / Feature Selection
You don’t have to train on all 365 features. Here are a few common strategies to reduce them:

###  Remove Near‑Constant Features
Drop any columns that are “almost always” 0 or 1 (very low variance).

```python
from sklearn.feature_selection import VarianceThreshold
target_cols = ['readmitted_<30', 'readmitted_>30', 'readmitted_NO']
# keep only features that vary at least 1% of the time
vt = VarianceThreshold(threshold=0.01)
X_reduced = vt.fit_transform(df.drop(columns=target_cols))
print("Kept", vt.get_support().sum(), "out of", df.shape[1], "features")
```

It seems that after applying the VarianceThreshold feature selection, you've reduced the dataset from 365 features to 135, meaning only features that have at least 1% variance were kept. This is a good approach to remove constant or near-constant features that don't provide much information for modeling.

```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# 1. Define your target columns (the dummies you don’t want in X)
target_cols = ['readmitted_<30', 'readmitted_>30', 'readmitted_NO']

# 2. Create X by dropping those target columns from your dataframe
X = df.drop(columns=target_cols)

# 3. (Optionally) inspect shapes
print("Before VarThresh:", X.shape)   # e.g. (289, 365)

# 4. Apply VarianceThreshold
vt = VarianceThreshold(threshold=0.01)   # keep features with ≥1% variance
X_reduced = vt.fit_transform(X)

# 5. Map back to column names
selected_columns = X.columns[vt.get_support()]

# 6. Build a reduced‐feature DataFrame
X_reduced_df = pd.DataFrame(X_reduced, columns=selected_columns)

# 7. Check results
print("After VarThresh:", X_reduced_df.shape)   # e.g. (289, 135)
print("Selected features:", list(selected_columns))

# Now X_reduced_df is your 289×135 feature matrix ready for modeling!
```
